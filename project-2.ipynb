```python
import  pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error ,mean_squared_error,r2_score
import matplotlib.pyplot as plt
import numpy as np

data =pd.read_csv("student.csv")

X=data[['Study_Hours']]
y=data['Total_Score']

model= LinearRegression()
model.fit(X,y)
predicted_score=model.predict(X)

mae= mean_absolute_error(y , predicted_score)
mse= mean_squared_error(y , predicted_score)
rmse = np.sqrt(mse)
r2=  r2_score(y,predicted_score)
print(round(mae,2))
print(round(mse,2))
print(round(rmse,4))
print(round(r2,5))
```

```python
plt.figure(figsize=(10,6))
plt.hist(data["Final_Score"],bins=30,color='skyblue',edgecolor='black')
plt.title("distribution of data")
plt.xlabel("Exam Score")
plt.ylabel("Number of Students ")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y,color='blue' ,label='Actual Score')
plt.plot(X,predicted_score,color="red",label="Prediced_score (regression_line)")

plt.title("model pred vs actual score")
plt.xlabel("study hours")
plt.ylabel("final output")
plt.grid(True)
plt.show()
```

```python
new_hours=9

predicted_new_score= model.predict([[new_hours]])
print(f"predicted new score for {new_hours} is {predicted_new_score}")
```
